## Feature Engineering

```python
df["zipcode"]=df["zipcode"].astype("object")
```

```python
import numpy as np
df['basement']= np.where(df['sqft_basement']>0, 1, 0)
df['renovated'] = np.where(df['yr_renovated']>0, 1, 0)
```

```python
df=df[df["bedrooms"]<11]
#10 dan kÃ¼Ã§Ã¼k yatak odalÄ±larÄ± dahil ediyoruz
```

```python
df["age"]=2022-df["yr_built"]
```

```python
df["bedrooms"]=df["bedrooms"]**2
#yatak odasÄ± sayÄ±sÄ±nÄ±n Ã¶nemini arttÄ±rÄ±r
```

```python
df["bathrooms"]=df["bathrooms"]**2
```

```python
df.quantile(.97).astype('int')
```

```python
outliers= df.quantile(.97)
df=df[(df["price"]<outliers["price"])]
df=df[(df["bedrooms"]<outliers["bedrooms"])]
df=df[(df["sqft_living"]<outliers["sqft_living"])]
```

```python
df=df[df["bathrooms"]<7]
```

## Modelling


```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
x=df[["sqft_living","grade","sqft_above","bathrooms","view"
      ,"bedrooms","zipcode","condition","waterfront","floors","renovated","age","lat","basement"]]
y=df[["price"]]
```

```python
x
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
#saat bazlÄ± random ilerlediÄinde hata alÄ±rÄ±z random state e bir deÄer atayÄ±p onun Ã¼zerinden gdersek daha doÄru sonuÃ§lar alÄ±rÄ±z
```

```python
model=lr.fit(x_train,y_train)
#modeli eÄitmek iÃ§in fit kullanÄ±lÄ±r
```

```python
tahmin=model.predict(x_test)
#eÄitilen model ile kenara saklanan verileri tahmin ederiz
```

```python
(mean_squared_error(tahmin,y_test))**0.5
#ne kadar hata yaptÄ±ÄÄ±mÄ±zÄ± gÃ¶sterir
```

```python
r2_score(tahmin,y_test)
```
